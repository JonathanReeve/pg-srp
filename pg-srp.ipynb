```python
import pandas as pd
from pySRP import SRP
import sqlite3
#import iteritems
#import six
```

```python
db = sqlite3.connect('../gitenberg-scrape/pg-text-7.db')
c = db.cursor()
```

```python
response = c.execute('select id from text')
listOfIds = response.fetchall()
```

```python
len(listOfIds)
```

```python
hasher = SRP.SRP(640)
```

```python
output = SRP.Vector_file("pg-srp-vecs.bin",dims=640,mode="w")
```

```python
def hashText(textId): 
    #print("Processing text: ", textId)
    #print(len(textId), type(textId), textId)
    response = c.execute('select text from text where id=?', [textId])
    text = response.fetchone()[0]
    print('Text: ', text.replace('\n', ' ')[:40])
    hashed = hasher.stable_transform(text)
    output.add_row(textId, hashed)
    #c.execute('')
    formatted = str(hashed.tolist())
    if len(formatted)<10: 
        print("Something went wrong with the hashing!")
    response2 = c.execute('update meta set srp=? where id=?', [formatted, str(float(textId))])
    db.commit()
    return response2
```

```python
start = 1
#11396
for bookIdPair in listOfIds[start:]:
    bookId = str(bookIdPair[0])
    print('Getting: ', bookId, type(bookId))
    hashText(bookId)
```
